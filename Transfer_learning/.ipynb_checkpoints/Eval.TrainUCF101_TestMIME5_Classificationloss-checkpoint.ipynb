#### Evaluation: Visualisation

```python
## Reading test data: Features

hf = h5py.File('/nobackup/leopauly/S2l3.0/Features_Storage/feature_trainUCF101_testmime5_fc6.h5', 'r')
feature_set_a=np.array(hf.get('feature_set_a'))
feature_set_b=np.array(hf.get('feature_set_b'))
feature_set_c=np.array(hf.get('feature_set_c'))
feature_set_d=np.array(hf.get('feature_set_d'))
feature_set_e=np.array(hf.get('feature_set_e'))
```

```python
print('Feature sizes:')
print(feature_set_a.shape)
print(feature_set_b.shape)
print(feature_set_c.shape)
print(feature_set_d.shape)
print(feature_set_e.shape)
```

```python
plt.rcParams.update({'font.size': 30})
plt.rcParams["font.family"] = "serif"
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams["figure.figsize"] = [16, 12]

## TSNE compression
tsne_obj = PCA(n_components=2)
points=np.concatenate((feature_set_a,feature_set_b,feature_set_c,feature_set_d,feature_set_e),axis=0)
print(points.shape)
tsne_obj.fit(points) 
```

```python
vis_tsne_a=tsne_obj.transform(feature_set_a) 
vis_tsne_a=np.array(vis_tsne_a)
print(tsne_obj.get_params())

vis_tsne_b=tsne_obj.transform(feature_set_b) 
vis_tsne_b=np.array(vis_tsne_b)
print(tsne_obj.get_params())

vis_tsne_c=tsne_obj.transform(feature_set_c) 
vis_tsne_c=np.array(vis_tsne_c)
print(tsne_obj.get_params())

vis_tsne_d=tsne_obj.transform(feature_set_d) 
vis_tsne_d=np.array(vis_tsne_d)
print(tsne_obj.get_params())

vis_tsne_e=tsne_obj.transform(feature_set_e) 
vis_tsne_e=np.array(vis_tsne_e)
print(tsne_obj.get_params())
```

```python
## Plotting
scatter_polt_size=350
plt.scatter(vis_tsne_a[:, 0], vis_tsne_a[:, 1],marker="X",s=scatter_polt_size,color='red',label='1')
plt.scatter(vis_tsne_b[:, 0], vis_tsne_b[:, 1],marker="v",s=scatter_polt_size,color='black',label='2')
plt.scatter(vis_tsne_c[:, 0], vis_tsne_c[:, 1],marker='o',s=scatter_polt_size,color='blue',label='3')
plt.scatter(vis_tsne_d[:, 0], vis_tsne_d[:, 1],marker="+",s=scatter_polt_size,color='orange',label='4')
plt.scatter(vis_tsne_e[:, 0], vis_tsne_e[:, 1],marker="x",s=scatter_polt_size,color='green',label='5')
plt.legend(bbox_to_anchor=(1.37, 1.025))
plt.show()
```

```python
# 3D plotting
fig = plt.figure(figsize=(100, 40))
ax = fig.add_subplot(251, projection='3d')
ax.scatter(vis_tsne_a[:, 0], vis_tsne_a[:, 1],marker="X",s=scatter_polt_size,color='red',label='1')
ax.scatter(vis_tsne_b[:, 0], vis_tsne_b[:, 1],marker="v",s=scatter_polt_size,color='black',label='2')
ax.scatter(vis_tsne_c[:, 0], vis_tsne_c[:, 1],marker='o',s=scatter_polt_size,color='blue',label='3')
ax.scatter(vis_tsne_d[:, 0], vis_tsne_d[:, 1],marker="+",s=scatter_polt_size,color='orange',label='4')
ax.scatter(vis_tsne_e[:, 0], vis_tsne_e[:, 1],marker="x",s=scatter_polt_size,color='green',label='5')
ax.view_init(20, 2)
plt.show()
```

```python
# 3D plotting
fig = plt.figure(figsize=(100, 40))
ax = fig.add_subplot(251, projection='3d')
ax.scatter(vis_tsne_a[:, 0], vis_tsne_a[:, 1],vis_tsne_a[:, 2],marker="X",s=scatter_polt_size,color='red',label='1')
ax.scatter(vis_tsne_b[:, 0], vis_tsne_b[:, 1],vis_tsne_b[:, 2],marker="v",s=scatter_polt_size,color='black',label='2')
ax.scatter(vis_tsne_c[:, 0], vis_tsne_c[:, 1],vis_tsne_c[:, 2],marker='o',s=scatter_polt_size,color='blue',label='3')
ax.scatter(vis_tsne_d[:, 0], vis_tsne_d[:, 1],vis_tsne_d[:, 2],marker="+",s=scatter_polt_size,color='orange',label='4')
ax.scatter(vis_tsne_e[:, 0], vis_tsne_e[:, 1],vis_tsne_e[:, 2],marker="x",s=scatter_polt_size,color='green',label='5')
ax.view_init(4, -72) #standard view: ax.view_init(4, -72)
plt.show()
```

#### Clustering Analysis

```python
kmeans = KMeans(n_clusters=num_class_test,random_state=1)
y_km = kmeans.fit_predict(points)
print(y_km.shape)
```

```python
print(kmeans)
```

```python
## Accuracy for our method
y_true_a=np.ones(feature_set_a.shape[0])*1
y_true_b=np.ones(feature_set_b.shape[0])*2
y_true_c=np.ones(feature_set_c.shape[0])*3
y_true_d=np.ones(feature_set_d.shape[0])*4
y_true_e=np.ones(feature_set_e.shape[0])*5
y_true=np.concatenate((y_true_a,y_true_b,y_true_c,y_true_d,y_true_e),axis=0)
print('Label shapes:')
print(y_true_a.shape)
print(y_true_b.shape)
print(y_true_c.shape)
print(y_true_d.shape)
print(y_true_e.shape)
print(y_true.shape)
```

```python
print('Adjusted random Index:',round(metrics.adjusted_rand_score(y_true,y_km),2))
```

```python
print('Adjusted mutual infromation score:',metrics.adjusted_mutual_info_score(y_true,y_km))
print('Homogeneity score:',metrics.homogeneity_score(y_true,y_km))
print('V measure score:',round(metrics.v_measure_score(y_true,y_km),2))
```
